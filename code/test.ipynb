```python
# used to load a pre-existing network and observe training metrics

from game import Game
from agent import Agent
import training_functions as tf
```

```python
version = "demo_version"

env = Game(size=4, 
           seed=42, 
           negative_reward=-2, 
           tile_move_penalty=0.1)

# starting with a low eps value for the pre-trained network to see how it is after training
eps = 0.00001

# Create the agent
agent = Agent(state_size=env.state_size * 18, 
              action_size=env.action_size,
              seed=42, 
              fc1_units=2048, 
              fc2_units=1024, 
              fc3_units=512,
              buffer_size=1000000, 
              batch_size=1024, 
              lr=0.00005, 
              use_expected_rewards=True, 
              predict_steps=2)

# Loading model weights
agent.load(version)
```

```python
tf.dqn(agent=agent, 
       env=env,
       version=version,
       n_episodes=200000,
       eps_start=eps,
       eps_end=0,
       eps_decay=0.999,
       start_learn_iterations=10,
       print_moves=False,
       stats_interval=100,
       saving_model=False)  # does not save model as to not overwrite original weights
```
